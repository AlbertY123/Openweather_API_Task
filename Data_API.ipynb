# Gathering data using an API

## Step 1: Get an API Key

Go to [OpenWeatherMap](https://openweathermap.org/appidand) sign up for a free account.

Once registered, navigate to the API keys section to generate your API key.

Copy the API key—you’ll need it to make requests.

## Step 2: Install Required Libraries

To interact with the API, you’ll need the requests library in Python. Install it using pip:

`pip install requests`

## Step 3: Make Your First API Request

Let’s start by fetching the current weather for a specific city.

```python
import requests
import json

# Replace with your API key
API_KEY = '5b8880c4f33c1388efc5c067473947c5'

# API endpoint and parameters
city = 'Melbourne'
url = f'http://api.openweathermap.org/data/2.5/weather?q={city}&appid={API_KEY}&units=metric'

# Make the request
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()  # Parse JSON response
    pretty_json_string = json.dumps(data, indent=4)
    print(pretty_json_string)  # Print the raw data
else:
    print(f"Error: {response.status_code}")
```

## Step 4: Parse and Use the Data

The API returns data in JSON format. Let’s extract and display some useful information:

```python
# Extract relevant data
if response.status_code == 200:
    data = response.json()
    city_name = data['name']
    weather_description = data['weather'][0]['description']
    temperature = data['main']['temp']  # Temperature in Kelvin
    humidity = data['main']['humidity']

    # Convert temperature to Celsius
    temperature_celsius = temperature

    # Display the data
    print(f"Weather in {city_name}:")
    print(f"Description: {weather_description}")
    print(f"Temperature: {temperature_celsius:.2f}°C")
    print(f"Humidity: {humidity}%")
```

```python
# Task 4
import requests
import json
import time

# Replace with your API key
API_KEY = '5b8880c4f33c1388efc5c067473947c5'

# API endpoint and parameters
city_1 = input("Enter the first city name: ")
url_1 = f'http://api.openweathermap.org/data/2.5/weather?q={city_1}&appid={API_KEY}&units=metric'

city_2 = input("Enter the second city name: ")
url_2 = f'http://api.openweathermap.org/data/2.5/weather?q={city_2}&appid={API_KEY}&units=metric'

# Make the request
response_city_1 = requests.get(url_1)
response_city_2 = requests.get(url_2)

# Check if the request was successful
if response_city_1.status_code == 200 and response_city_2.status_code == 200:
    data_1 = response_city_1.json()  # Parse JSON response for first city
    data_2 = response_city_2.json()  # Parse JSON response for second city
    
    #Compare weather of 2 city
    print(f"Comparing weather between {city_1} and {city_2}...")
    time.sleep(1)  # Add a delay between comparisons
    #Temperature of cities
    temp_city_1 = data_1['main']['temp']
    temp_city_2 = data_2['main']['temp']
    print(f"Temperature in {city_1}: {temp_city_1:.2f}°C")
    print(f"Temperature in {city_2}: {temp_city_2:.2f}°C")
    if temp_city_1 > temp_city_2:
        print(f"{city_1} is warmer than {city_2} with a temperature of {temp_city_1:.2f}°C compared to {temp_city_2:.2f}°C.")
    elif temp_city_1 < temp_city_2:
        print(f"{city_2} is warmer than {city_1} with a temperature of {temp_city_2:.2f}°C compared to {temp_city_1:.2f}°C.")
    else:
        print(f"Both {city_1} and {city_2} have the same temperature of {temp_city_1:.2f}°C.")
    time.sleep(1)  # Add a delay between comparisons
    #Humidity of cities
    humidity_city_1 = data_1['main']['humidity']
    humidity_city_2 = data_2['main']['humidity']
    
    print(f"Humidity in {city_1}: {humidity_city_1}%")
    print(f"Humidity in {city_2}: {humidity_city_2}%")
    if humidity_city_1 > humidity_city_2:
        print(f"{city_1} is more humid than {city_2} with a humidity of {humidity_city_1}% compared to {humidity_city_2}%.")
    elif humidity_city_1 < humidity_city_2:
        print(f"{city_2} is more humid than {city_1} with a humidity of {humidity_city_2}% compared to {humidity_city_1}%.")
    else:
        print(f"Both {city_1} and {city_2} have the same humidity of {humidity_city_1}%.")

else:
    print("There was an error with the API request. Please check the city name and try again.")
```
